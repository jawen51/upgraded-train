```python
for i in range(0,3):
    filename = "Perf_"+str(i).zfill(2) + ".zip"
    #print(filename)
    temp = pd.read_csv(filename, compression='zip', sep=',', encoding='gbk')
    print(i)
    temp["hour_id"] = i
    temp = pd.DataFrame(temp)
    Perf = pd.concat([Perf,temp],ignore_index= True)
```
